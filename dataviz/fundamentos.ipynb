###Matplotlib

```python
import matplotlib.pyplot as plt
```

####Criando um gráfico de linha

```python
#Dados fictícios - Vendas ao longo do meses
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
vendas = [150, 200, 180, 300, 250, 400]
```

```python
#Criando o gráfico de linha
plt.figure(figsize=(8, 5))
plt.plot(
    meses,
    vendas,
    marker='o',
    linestyle='-',
    color='blue',
    label='Vendas'
)
plt.show()
```

```python
#Adicionando rótulos e título ao gráfico
plt.figure(figsize=(8, 5))
plt.plot(
    meses,
    vendas,
    marker='o',
    linestyle='-',
    color='blue',
    label='Vendas'
)
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.title('Vendas ao longo dos meses')
plt.legend()
plt.grid(True)
plt.show()
```

####Criando um gráfico de barra

```python
#Dados fictícios - Quantidade de produtos vendidos por vendedor
vendedores = ['João', 'Maria', 'Pedro', 'Ana']
quantidade_vendidas = [45, 60, 30, 55]
```

```python
#Criando gráfico de barras
plt.figure(figsize=(8,5))
plt.bar(
    vendedores,
    quantidade_vendidas,
    color='green'
)
plt.show()
```

```python
#Adiconando rótulos e título ao gráfico
plt.figure(figsize=(8,5))
plt.bar(
    vendedores,
    quantidade_vendidas,
    color='green',
)

plt.xlabel('vendedores')
plt.ylabel('quantidade Vendidas')
plt.title('Quantidade de Produtos Vendidos por Vendedor')
plt.show()
```

###Criando gráfico de dispersão

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Gerando dados aleatórios
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
print(x)
print(y)
```

```python
#Criando o gráfico de dispersão
plt.figure(figsize=(8,6))
plt.scatter(x,y)
plt.title('Gráfico de Dispersão com Dados Aleatórios')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.grid(True)
plt.show()
```

```python
#Criando o gráfico 3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(x,y,z)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
ax.set_title('Gráfico de dispersão em 3D')
plt.show()
```

###Gráfico Histograma

```python
#Gerando dados fictícios - Pontuação de um teste
pontuacoes = np.random.randint(50, 100, 100)
```

```python
#Criando o histograma
plt.figure(figsize=(8,5))
plt.hist(
    pontuacoes,
    bins=10,
    color='skyblue',
    edgecolor='black'
)

plt.xlabel('Pontuação')
plt.ylabel('Frequência')
plt.title('Distribuição das pontuações do teste')

plt.show()
```
