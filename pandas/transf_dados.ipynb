```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('dados_hospedagem.json')
```

```python
df.head()
```

```python
df = pd.json_normalize(df['info_moveis'])
```

```python
df.head()
```

```python
columns = list(df.columns)
columns
```

```python
df = df.explode(columns[3:])
df.head()
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df.head()
```

```python
df.dtypes
```

Conversão dados numéricos

```python
df['max_hospedes'] = df['max_hospedes'].astype(np.int64)
```

```python
df.dtypes
```

```python
columns_numerics = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
df[columns_numerics] = df[columns_numerics].astype(np.int64)
```

```python
df.dtypes
```

```python
df['avaliacao_geral']
```

```python
df['avaliacao_geral'] = df['avaliacao_geral'].astype(np.float64)
```
