```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('dados_hospedagem.json')
```

```python
df.head()
```

```python
df = pd.json_normalize(df['info_moveis'])
```

```python
df.head()
```

```python
columns = list(df.columns)
columns
```

```python
df = df.explode(columns[3:])
df.head()
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df.head()
```

```python
df.dtypes
```

Conversão dados numéricos

```python
df['max_hospedes'] = df['max_hospedes'].astype(np.int64)
```

```python
df.dtypes
```

```python
columns_numerics = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
df[columns_numerics] = df[columns_numerics].astype(np.int64)
```

```python
df.dtypes
```

```python
df['avaliacao_geral']
```

```python
df['avaliacao_geral'] = df['avaliacao_geral'].astype(np.float64)
```

```python
df.dtypes
```

```python
df['preco']
```

```python
df['preco'] = df['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
df['preco']
```

```python
df['preco'] = df['preco'].astype(np.float64)
```

```python
df.dtypes
```

```python
df[['taxa_deposito', 'taxa_limpeza']]
```

```python
df[['taxa_deposito', 'taxa_limpeza']] = df[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
df[['taxa_deposito', 'taxa_limpeza']] = df[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
df.dtypes
```

Conversão dados textuais 

```python
df['descricao_local']
```

```python
df['descricao_local'] = df['descricao_local'].str.lower()
df['descricao_local']
```

```python
df['descricao_local'][3000]
```

```python
df['descricao_local'] = df['descricao_local'].str.replace('[^a-zA-Z0-9\-\*]', ' ', regex=True)
```

```python
df['descricao_local'][3000]
```

```python
df['descricao_local'] = df['descricao_local'].str.replace('(?<!\w)', ' ',regex=True)
```

```python
df.head()
```

```python
df['descricao_local'] = df['descricao_local'].str.split()
df
```

```python
df['comodidades']
```

```python
df['comodidades'] = df['comodidades'].str.replace('\{|}|\´', '', regex=True)
```

```python
df['comodidades']
```

```python
df['comodidades'] = df['comodidades'].str.split(',')
```
